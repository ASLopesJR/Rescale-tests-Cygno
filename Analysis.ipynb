```python
import matplotlib
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
from pathlib import Path
```

```python
pathstring = './moved_means/noise/'
path  = Path(pathstring)
noise_var2  = np.array(0)
noise_var4  = np.array(0)
noise_var8  = np.array(0)
noise_var16 = np.array(0)
noise_var32 = np.array(0)

for entry in path.iterdir():
    if entry.name.split('_')[-1] == '2.txt':
        meanfr = np.loadtxt(pathstring+entry.name)
        maxvalues = np.max(meanfr,1)
        noise_var2 = np.append(noise_var2,maxvalues)
    elif entry.name.split('_')[-1] == '4.txt':
        meanfr = np.loadtxt(pathstring+entry.name)
        maxvalues = np.max(meanfr,1)
        noise_var4 = np.append(noise_var4,maxvalues)
    elif entry.name.split('_')[-1] == '8.txt':
        meanfr = np.loadtxt(pathstring+entry.name)
        maxvalues = np.max(meanfr,1)
        noise_var8 = np.append(noise_var8,maxvalues)
    elif entry.name.split('_')[-1] == '16.txt':
        meanfr = np.loadtxt(pathstring+entry.name)
        maxvalues = np.max(meanfr,1)
        noise_var16 = np.append(noise_var16,maxvalues)
    elif entry.name.split('_')[-1] == '32.txt':
        meanfr = np.loadtxt(pathstring+entry.name)
        maxvalues = np.max(meanfr,1)
        noise_var32 = np.append(noise_var32,maxvalues)
```
